```
#Funcion para hallar la distancia de cada nodo en la red de panqueques
def a(key, listo, distanci, nums):
        if(listo[key]):
           distanci[key] += 1 #Aumenta la distancia entre la permutacion y el nodo organizado
           for i in listo[key]:
                a(i, listo, distanci, nums)
        else:
            return(distanci)

#Funcion para hallar el Pn 
def xd(lista):
    dist={}  #Diccionario que contiene las distancias de cada nodo
    for x in lista:
        dist[x]=0
        if lista[x]:
            dist[x] += 1
    nums=[]
    ad = {}
    for x in lista:
        for y in lista[x]:
            ad = a(y, lista, dist, nums)
    all_values = ad.values()
    try:
        print(max(all_values))
    except:
        print("0")


    return 0
```

```
def ModifiedBFSRed(s): 

        # Create dictionaries graph,visited and pydot graph G
        graph = defaultdict()
        visited = defaultdict()
        #Grafo no dirigido ya que estamos haciendo la red
        #Lo ponemos estricto para que solo halla un camino entre cada nodo
        G = pydot.Dot(graph_type="graph", strict=True)  
        
        # Create a queue for BFS 
        queue = [] 
  
        # Mark the source node as visited and enqueue it 
        queue.append(s) 
        visited[lstr(s)] = True
  
        while queue: 
            # Dequeue a vertex from queue 
            #print("queue", queue)
            s = queue.pop(0) 
            graph[lstr(s)]=[]
            node = pydot.Node(lstr(s))
            G.add_node(node)
            
            # Get all adjacent vertices of the 
            # dequeued vertex s. If a adjacent 
            # has not been visited, then mark it 
            # visited and enqueue it 
            if(s != sorted(s)):
              for i in range(len(s)-1):
                d = s[i+1::-1] + s[i+2:]
                #print("s",s)
                #print("d",d)
                if (d != s) and not(lstr(d) in visited.keys()):
                 
                  if(d != sorted(d)):
                    node = pydot.Node(lstr(d))
                  else:
                    node = pydot.Node(lstr(d),style="filled", fillcolor="green") # first sorted green
                  
                  G.add_node(node)
                  edge = pydot.Edge(lstr(s),lstr(d))
                  G.add_edge(edge)
                  graph[lstr(s)].append(lstr(d))
                  queue.append(d) 
                  visited[lstr(d)] = True 
                else:
                    edge = pydot.Edge(lstr(s),lstr(d)) #Para completar la red
                    G.add_edge(edge)
        Pn = 0
        dist = {}       
        #Para calcular Pn
        xd(graph)
                                           
        # Dsplay the pydot graph G
        im = Image(G.create_png())
        display(im)
```

<h2><b>3. Modify the PancakesTrees.ipynb code to plot the pancakes trees using two spatulas.

```
def ModifiedBFS2Spat(s): 

        # Create dictionaries graph,visited and pydot graph G
        graph = defaultdict()
        visited = defaultdict()
        G = pydot.Dot(graph_type="digraph")
        
        # Create a queue for BFS 
        queue = [] 
  
        # Mark the source node as visited and enqueue it 
        queue.append(s) 
        visited[lstr(s)] = True
  
        while queue: 
  
            # Dequeue a vertex from queue 
            s = queue.pop(0) 
            graph[lstr(s)]=[]
            node = pydot.Node(lstr(s))
            G.add_node(node)
            
            # Get all adjacent vertices of the 
            # dequeued vertex s. If a adjacent 
            # has not been visited, then mark it 
            # visited and enqueue it 
            if(s != sorted(s)):
              for i in range(len(s)-1): 
                  for j in range(i+1):          #Agregamos un for mas para recojer todos los posibles flips
                    d=s[0:j]+s[j:i+2][::-1]+s[i+2:]  #Cambiamos el generador para hacer el flip en los lugar establecido
                    if (d != s) and not(lstr(d) in visited.keys()):
                    
                        if(d != sorted(d)):
                            node = pydot.Node(lstr(d))
                        else:
                            node = pydot.Node(lstr(d),style="filled", fillcolor="green") # first sorted green
                        
                        G.add_node(node)
                        edge = pydot.Edge(lstr(s),lstr(d))
                        G.add_edge(edge)
                        graph[lstr(s)].append(lstr(d))
                        queue.append(d) 
                        visited[lstr(d)] = True 
              
        # Dsplay the pydot graph G
        im = Image(G.create_png())
        display(im)
```
