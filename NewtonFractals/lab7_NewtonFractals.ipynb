```
import math
import matplotlib.pyplot as plt
import numpy as np

def f(z):
    return z ** 8 + 15*z**4 - 16

def df(z):
    return 8*z**7 + 60*z**3

def newton_set1(z_c_r=0, z_c_i=0, zh = 5, zw=5, width=500, height=500, zoom=1,  niter=256):
    """ Fractals using newton-raphson """

    # Pixels array
    pixels = np.arange(width*height*3, dtype=np.uint32).reshape(height, width, 3)

    h = 1e-7 # step size for numerical derivative
    eps = 1e-3 # max error allowed

    # Bounding roots
    r1 = 1
    r2=-1
    r3 = complex(0,-1)
    r4 = complex(0, 1)
    r5 = complex(-math.sqrt(2), math.sqrt(2))
    r6 = complex(-math.sqrt(2), -math.sqrt(2))
    r7 = complex(math.sqrt(2), math.sqrt(2))
    r8 = complex(math.sqrt(2), -math.sqrt(2))

    # Color multiplication factor
    # NOTE: Increasing this darkens the image and lightens it otherwise
    multcol =5

    for y in range(height):
            zy =   z_c_i+(((y-(height/2))/height)*zh*zoom)

            for x in range(width):
                zx = z_c_r+ (((x-(width/2))/width)*zw*zoom)

                # Mapping real -> complex plane
                z = complex(zx, zy)
                
                count = 0
                
                for i in range(niter):
                    # complex numerical derivative
                    dz = df(z)
                    if dz == 0:
                        break

                    count += 1
                    if count > 255:
                        break
                   
                    znext = z - f(z) / dz # Newton iteration

                    if abs(znext - z) < eps: # stop when close enough to any root
                        break
                    
                    z = znext

                # Pixels colored using the roots
                if abs(z-r1)<eps:
                    # color red
                    pixels[height-y-1,x] = (255 - count*multcol, 0, 0)
                elif abs(z-r5)<=eps:
                    # color green
                    pixels[height-y-1,x] = (255- count*multcol, 0 , 227- count*multcol)
                elif abs(z-r3)<=eps:
                    # color blue
                    pixels[height-y-1,x] = (0, 0, 255 - count*multcol)
                elif abs(z-r4)<=eps:
                    # color blue
                    pixels[height-y-1,x] = (255- count*multcol, 179 - count*multcol, 0 )
                elif abs(z-r8)<=eps:
                    # color blue
                    pixels[height-y-1,x] = (255- count*multcol, 254 - count*multcol, 0)
                elif abs(z-r6)<=eps:
                    # color blue
                    pixels[height-y-1,x] = (97 - count*multcol, 255- count*multcol, 0)
                elif abs(z-r7)<=eps:
                    # color blue
                    pixels[height-y-1,x] = (0, 255- count*multcol, 197 - count*multcol)
                elif abs(z-r2)<=eps:
                    # color blue
                    pixels[height-y-1,x] = (164- count*multcol, 0, 255 - count*multcol)
                else:   
                    pixels[height-y-1,x] = (0, 0, 0)
 
    return pixels

def display1(z_c_r=0, z_c_i=0, zh = 1, zw=1, width=500, height=500, zoom=1,  niter=256):
    """ Display a newton-raphson fractal """
    print("Z region from (",z_c_r+ (((0-(width/2))/width)*zw*zoom),",",z_c_i+(((0-(height/2))/height)*zh*zoom) ,") to (",z_c_r+ (((width-(width/2))/width)*zw*zoom),",",z_c_i+(((height-(height/2))/height)*zh*zoom),")")
    pimg = newton_set1(z_c_r=z_c_r, z_c_i=z_c_i, zh = zh, zw=zw, width=width, height=width, zoom=zoom,  niter=niter)
    plt.title('Newton Fractal f(z)=z**3-1')
    plt.xlabel("real"+"("+ str(z_c_r+ (((0-(width/2))/width)*zw*zoom))+","+str(z_c_r+ (((width-(width/2))/width)*zw*zoom))+")")
    plt.ylabel("imaginary"+"("+str(z_c_i+(((0-(height/2))/height)*zh*zoom))+","+str(z_c_i+(((height-(height/2))/height)*zh*zoom)) +")")
    plt.imshow(pimg)
    plt.show()
```

```
print("zoom out")
display1(z_c_r=0.9, z_c_i=0.9,zoom=1)
display1(zoom=2)
display1(zoom=2.5)
display1(z_c_r=-1.5, z_c_i=1.5,zoom=4)
display1(zoom=5)
display1(zoom=7)
display1(zoom=8)
display1(zoom=9)
display1(zoom=10)
```

#Anlisis del fractal de Newton para $ x^8+15x^4-16$# 


<div style="text-align:justify">En primer lugar se obtuvo un fractal simtrico con 8 regiones amplias de convergencia, estas regiones respectivamente convergen a las 8 raices del polinomio, y se observa que en puntos cercanos al origen converge a distintas raices en zonas ms pequeas, y este patron se repite en el contorno de las regiones amplias de convergencia de cada raz. A si mismo en cada uno de los ejes tanto el positivo y negativo en "y" como en "x", por estos ejes a medida que los puntos se acercan al origen tienden a converger mas rapido a 4 de las 8 raices estas zonas son las de color naranja,azul,rojo,morado; sin embargo en cierto punto muy cercano al origen, ya tiende a converger a raices distintas en zonas pequeas. Por otro lado en las diagonales de cada cuadrante cuando los puntos se acercan por estas diagonales al origen, converge ms rapido a las otras 4 raices de color verde,amarillo,celeste,rosado; sin embargo llega a un punto en esa diagonal donde converge a distintas raices en zonas pequeas y as sigue hasta el origen.
</div>


```
import math
import matplotlib.pyplot as plt
import numpy as np

def f(z):
    return z ** 6 + z ** 3 - 1

def df(z):
    return 6*z**5 + 3*z**2

def newton_set1(z_c_r=0, z_c_i=0, zh = 5, zw=5, width=500, height=500, zoom=1,  niter=256):
    """ Fractals using newton-raphson """

    # Pixels array
    pixels = np.arange(width*height*3, dtype=np.uint32).reshape(height, width, 3)

    h = 1e-7 # step size for numerical derivative
    eps = 1e-3 # max error allowed

    # Bounding roots
    r1 = -(1/2 + math.sqrt(5)/2)**(1/3)
    r2= (np.sqrt(5)/2 - 1/2)**(1/3)
    r3 = complex(((1/2)*(-1 - math.sqrt(5)))**(1/3))
    r4 = complex(-(-1/2)**(1/3)*(-1 - math.sqrt(5))**(1/3))
    r5 = complex(-((1/2)*(1 - math.sqrt(5)))**(1/3))
    r6 = complex((-1)**(2/3)*((1/2)*(math.sqrt(5) - 1))**(1/3))
    

    # Color multiplication factor
    # NOTE: Increasing this darkens the image and lightens it otherwise
    multcol =5
    diverge = 0
    total = 0
    for y in range(height):
            zy =   z_c_i+(((y-(height/2))/height)*zh*zoom)

            for x in range(width):
                zx = z_c_r+ (((x-(width/2))/width)*zw*zoom)

                # Mapping real -> complex plane
                z = complex(zx, zy)
                
                count = 0
                
                for i in range(niter):
                    # complex numerical derivative
                    dz = df(z)
                    if dz == 0:
                        break

                    count += 1
                    if count > 255:
                        break
                   
                    znext = z - f(z) / dz # Newton iteration

                    if abs(znext - z) < eps: # stop when close enough to any root
                        break
                    
                    z = znext

                # Pixels colored using the roots
                total += 1
                if abs(z-r1)<eps:
                    # color red
                    pixels[height-y-1,x] = (239, 17 + count*multcol, 5)
                elif abs(z-r5)<=eps:
                    # color negro
                    pixels[height-y-1,x] = (count*multcol,count*multcol,count*multcol)
                elif abs(z-r3)<=eps:
                    # color gris
                    pixels[height-y-1,x] = (100- 0.5*count*multcol, 100-0.5*count*multcol, 100-0.5*count*multcol)
                    
                elif abs(z-r4)<=eps:
                    # color verde limn
                    pixels[height-y-1,x] = (205- count*multcol, 190,4 )
                    
                elif abs(z-r6)<=eps:
                    # color verde fluorescente
                    pixels[height-y-1,x] = (150 - count*multcol, 255- count*multcol, 0)
                    
                elif abs(z-r2)<=eps:
                    # color amarillo
                    pixels[height-y-1,x] = (241, 237 - count*multcol, 6)
                    
                else:   
                    pixels[height-y-1,x] = (0, 0, 0)
                    diverge += 1
                    #print("--")
                    #print(height-y-1,x)
                    #print("..")
                    #print(z)
                    #print("--")
    print("DIVERGE", diverge)
    print("TOTAL", total)
    return pixels

def display1(z_c_r=0, z_c_i=0, zh = 1, zw=1, width=500, height=500, zoom=1,  niter=256):
    """ Display a newton-raphson fractal """
    print("Z region from (",z_c_r+ (((0-(width/2))/width)*zw*zoom),",",z_c_i+(((0-(height/2))/height)*zh*zoom) ,") to (",z_c_r+ (((width-(width/2))/width)*zw*zoom),",",z_c_i+(((height-(height/2))/height)*zh*zoom),")")
    pimg = newton_set1(z_c_r=z_c_r, z_c_i=z_c_i, zh = zh, zw=zw, width=width, height=width, zoom=zoom,  niter=niter)
    plt.title('Newton Fractal f(z)=z**6 + z**3 - 1')
    plt.xlabel("real"+"("+ str(z_c_r+ (((0-(width/2))/width)*zw*zoom))+","+str(z_c_r+ (((width-(width/2))/width)*zw*zoom))+")")
    plt.ylabel("imaginary"+"("+str(z_c_i+(((0-(height/2))/height)*zh*zoom))+","+str(z_c_i+(((height-(height/2))/height)*zh*zoom)) +")")
    
    plt.imshow(pimg)
    plt.show()
```

#Anlisis del fractal de Newton para $ x^6+x^3-1$# 


<div style="text-align:justify">
Se obtuvo un fractal simtrico con respecto a una recta con pendiente negativa que pasa por el centro del mismo. El fractal consiste de 6 regiones amplias de convergencia, las cuales convergen a las 6 raices del polinomio. En las divisiones entre regiones, la existe una convergencia a distintas races en una zona ms pequea. Existen 3 races:
<li> r1 = $-(\frac{1}{2} + \frac{\sqrt{5}}{2})^{\frac{1}{3}}$ zona roja
<li>r3 = $((\frac{1}{2})(-1 - \sqrt{5}))^{\frac{1}{3}}$ zona gris
<li>r4 = $(-(\frac{-1}{2})^{\frac{1}{3}}*(-1 - \sqrt{5})^{\frac{1}{3}})$ zona verde limn<br> que al acercarse al centro, convergen a la raiz amarilla, la raiz negra y a la raiz verde fluorescente correspondientemente. En las divisiones entre regiones tambin tienen dichos colores para las regiones de races listadas.
</div>


```
def f(z):
    return cmath.sin(z)

def df(z):
    try: 
      return cmath.cos(z)
    except:
      return 1

def newton_set1(z_c_r=0, z_c_i=0, zh = 5, zw=5, width=500, height=500, zoom=1,  niter=256):
    """ Fractals using newton-raphson """

    # Pixels array
    pixels = np.arange(width*height*3, dtype=np.uint32).reshape(height, width, 3)

    h = 1e-7 # step size for numerical derivative
    eps = 1e-3 # max error allowed

    # Bounding roots
    r1 = 0
    r2 = math.pi
    r4 = 2*math.pi
    r5 = 3*math.pi
    r6 = 4*math.pi
    r7 = 5*math.pi
    r8 = 6*math.pi
    r9 = -math.pi
    r10 = -2*math.pi
    r11 = -3*math.pi
    r12 = -4*math.pi
    r13 = -5*math.pi
    r14 = -6*math.pi
    r15 = -7*math.pi
    r16 = 7*math.pi
    

    # Color multiplication factor
    # NOTE: Increasing this darkens the image and lightens it otherwise
    multcol = 3

    for y in range(height):
            zy =   z_c_i+(((y-(height/2))/height)*zh*zoom)

            for x in range(width):
                zx = z_c_r+ (((x-(width/2))/width)*zw*zoom)

                # Mapping real -> complex plane
                z = complex(zx, zy)
                
                count = 0
                
                for i in range(niter):
                    # complex numerical derivative
                    dz = df(z)
                    if dz == 0:
                        break

                    count += 1
                    if count > 255:
                        break
                   
                    try:
                      znext = z - f(z) / cmath.cos(z) # Newton iteration
                    except:
                      break

                    if abs(znext - z) < eps: # stop when close enough to any root
                        break
                    
                    z = znext

                # Pixels colored using the roots
                if abs(z-r1)<eps:
                    # color red
                    pixels[height-y-1,x] = (255 - count*multcol, 0, 0)
                elif abs(z-r2)<=eps:
                    # color green
                    pixels[height-y-1,x] = (0, 0, 255 - count*multcol)
                elif abs(z-r4)<=eps:
                    # color blue
                    pixels[height-y-1,x] = (0 , 85, 255- count*multcol)
                elif abs(z-r5)<=eps:
                    # color blue
                    pixels[height-y-1,x] = (0, 200, 255- count*multcol)
                elif abs(z-r6)<=eps:
                    # color blue
                    pixels[height-y-1,x] = (0, 255, 255 - count*multcol)
                elif abs(z-r7)<=eps:
                    # color blue
                    pixels[height-y-1,x] = (0, 255, 188 - count*multcol)
                elif abs(z-r8)<=eps:
                    # color blue
                    pixels[height-y-1,x] = (0, 170, 111 - count*multcol)
                elif abs(z-r9)<=eps:
                    # color blue
                    pixels[height-y-1,x] = (255 - count*multcol, 100, 1)
                elif abs(z-r10)<=eps:
                    # color blue
                    pixels[height-y-1,x] = (255 - count*multcol, 150, 10)
                elif abs(z-r11)<=eps:
                    # color blue
                    pixels[height-y-1,x] = (255 - count*multcol, 200, 20)
                elif abs(z-r12)<=eps:
                    # color blue
                    pixels[height-y-1,x] = (255 - count*multcol, 225, 0)
                elif abs(z-r13)<=eps:
                    # color blue
                    pixels[height-y-1,x] = (200 - count*multcol, 255, 0)
                elif abs(z-r14)<=eps:
                    # color blue
                    pixels[height-y-1,x] = (173 - count*multcol, 200, 0)
                elif abs(z-r15)<=eps:
                    # color blue
                    pixels[height-y-1,x] = (140 - count*multcol, 200, 0)
                elif abs(z-r16)<=eps:
                    # color blue
                    pixels[height-y-1,x] = (0, 206, 11 - count*multcol)
                else:   
                    pixels[height-y-1,x] = (255, 255, 255)
 
    return pixels

def display1(z_c_r=0, z_c_i=0, zh = 1, zw=1, width=500, height=500, zoom=1,  niter=256):
    """ Display a newton-raphson fractal """
    print("Z region from (",z_c_r+ (((0-(width/2))/width)*zw*zoom),",",z_c_i+(((0-(height/2))/height)*zh*zoom) ,") to (",z_c_r+ (((width-(width/2))/width)*zw*zoom),",",z_c_i+(((height-(height/2))/height)*zh*zoom),")")
    pimg = newton_set1(z_c_r=z_c_r, z_c_i=z_c_i, zh = zh, zw=zw, width=width, height=width, zoom=zoom,  niter=niter)
    plt.title('Newton Fractal f(z)=sen(z)')
    plt.xlabel("real"+"("+ str(z_c_r+ (((0-(width/2))/width)*zw*zoom))+","+str(z_c_r+ (((width-(width/2))/width)*zw*zoom))+")")
    plt.ylabel("imaginary"+"("+str(z_c_i+(((0-(height/2))/height)*zh*zoom))+","+str(z_c_i+(((height-(height/2))/height)*zh*zoom)) +")")
    plt.imshow(pimg)
    plt.show()
```

Este fractal converge a raices infinitas, por eso cuando lo alejamos se pueden observar las franjas de colores, aqui la parte blanca indica divergencia, pero el fractal no diverge en estos puntos, solo que tendramos que poner colores infinitos para marcar estas franjas.
